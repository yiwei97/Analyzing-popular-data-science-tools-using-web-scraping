```python
# Read packages
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Read files
ds_jobs_cleaned = pd.read_csv('../2_Data_cleaning/Output/job_offers_cleaned.csv')
```

```python
# Filter out relevant rows with 'data' in title
ds_jobs = ds_jobs_cleaned.loc[ds_jobs_cleaned['job_title'].str.contains('Data'),:]

# Filter out internship roles
ds_jobs['job_title_intern'] = ds_jobs['job_title'].str.lower().str.contains('intern', na=True).astype(int)
ds_jobs['work_type_intern'] = ds_jobs['work_type'].str.lower().str.contains('intern', na=True).astype(int)
ds_jobs['seniority_level_intern'] = ds_jobs['seniority_level'].str.lower().str.contains('intern', na=True).astype(int)
ds_jobs['intern'] = ds_jobs['job_title_intern'] + ds_jobs['work_type_intern'] + ds_jobs['seniority_level_intern']
ds_jobs['intern'] = np.clip(ds_jobs['intern'], a_max = 1, a_min = None)
ds_jobs = ds_jobs.drop(['job_title_intern', 'work_type_intern', 'seniority_level_intern'], axis=1)

# Filter out non-internship roles
ds_jobs = ds_jobs.loc[ds_jobs['intern']==0,:]

# Filter out only 'data scientist' or 'data analyst'
ds_jobs = ds_jobs.loc[ds_jobs['job_title'].str.lower().str.contains('data scien|data analyst', regex=True),:]
```

```python
# Visualize data: Jobs needed by company size
ds_by_company_size = pd.DataFrame(ds_jobs.groupby('no_of_employees')['no_of_employees'].count())
ds_by_company_size = ds_by_company_size.rename(columns = {ds_by_company_size.index.name:'count'}).reset_index()
ds_by_company_size['helper'] = ds_by_company_size['no_of_employees'].str.len()
ds_by_company_size.loc[ds_by_company_size['no_of_employees']=="10001+",'helper'] = 100
ds_by_company_size = ds_by_company_size.sort_values('helper').drop(['helper'], axis=1)

# Creating the bar plot
plt.bar(ds_by_company_size['no_of_employees'], ds_by_company_size['count'], color ='maroon',
        width = 0.4)
plt.xticks(rotation='vertical') 
plt.xlabel("No. of employees")
plt.ylabel("No. of job postings")
plt.title("Job postings by company size")
plt.show()
```

```python
# Visualize data: Jobs needed by industry
ds_by_industry = pd.DataFrame(ds_jobs.groupby('industry_new')['industry_new'].count())
ds_by_industry = ds_by_industry.rename(columns = {ds_by_industry.index.name:'count'}).reset_index()
ds_by_industry = ds_by_industry.sort_values('count', ascending=False)

# Creating the bar plot
plt.bar(ds_by_industry['industry_new'], ds_by_industry['count'], color ='maroon',
        width = 0.4)
plt.xticks(rotation='vertical') 
plt.xlabel("Industry")
plt.ylabel("No. of job postings")
plt.title("Job postings by industry")
plt.show()
```

```python
# Visualize tools needed
ds_by_tools = pd.DataFrame(ds_jobs.agg({'tool_python':'sum', 'tool_r':'sum','tool_sql':'sum'}))
ds_by_tools = ds_by_tools.reset_index().rename(columns={'index':'tools', 0:'count'})
ds_by_tools['tools'] = ds_by_tools['tools'].str.split('_', expand=True)[1].str.title()
ds_by_tools = ds_by_tools.sort_values('count', ascending = False)

# Creating the bar plot
plt.bar(ds_by_tools['tools'], ds_by_tools['count'], color ='maroon',
        width = 0.4)
plt.xticks(rotation='vertical') 
plt.xlabel("Type of tools")
plt.ylabel("No. of job postings")
plt.title("No. of job postings by type of tools")
plt.show()
```

```python
# Visualize libraries needed
ds_by_libraries = pd.DataFrame(ds_jobs[['lib_numpy', 'lib_matplotlib',
       'lib_pandas', 'lib_scipy', 'lib_scikit', 'lib_theano', 'lib_tensorflow',
       'lib_keras', 'lib_pytorch', 'lib_cntk', 'lib_mxnet']].agg('sum'))
ds_by_libraries = ds_by_libraries.reset_index().rename(columns={'index':'libraries', 0:'count'})
ds_by_libraries['libraries'] = ds_by_libraries['libraries'].str.split('_', expand=True)[1].str.title()
ds_by_libraries = ds_by_libraries.sort_values('count', ascending = False)

# Creating the bar plot
plt.bar(ds_by_libraries['libraries'], ds_by_libraries['count'], color ='maroon',
        width = 0.4)
plt.xticks(rotation='vertical') 
plt.xlabel("Type of libraries")
plt.ylabel("No. of job postings")
plt.title("No. of job postings by type of libraries")
plt.show()
```

```python
# Visualize job postings by date to understand how long vacancies take to be filled
ds_by_post_date = ds_jobs.copy()
ds_by_post_date['count'] = "temp"
ds_by_post_date['days_from_post_date'] = np.clip(ds_by_post_date['days_from_post_date'], a_max = 100, a_min = 7)
ds_by_post_date = pd.DataFrame(ds_by_post_date.groupby('days_from_post_date')['count'].count())
date_map = {'7':'1 week',
           '14':'2 weeks',
           '21':'3 weels',
           '28':'1 month',
           '56':'2 months'}
ds_by_post_date = ds_by_post_date.reset_index()
ds_by_post_date['duration'] = ds_by_post_date['days_from_post_date'].astype(str).map(date_map)

# Creating the bar plot
plt.bar(ds_by_post_date['duration'], ds_by_post_date['count'], color ='maroon',
        width = 0.4)
plt.xticks(rotation='vertical') 
plt.xlabel("Duration from job posting")
plt.ylabel("No. of job postings")
plt.title("No. of job postings by duration")
plt.show()
```

```python
# Visualize number of applicants by company_size to understand if applicants select based on company size
ds_by_applicants = ds_jobs.copy()
employees_map = {'501-1000':'501-1000', 
                 '10001+':'1001+', 
                 '1001-5000':'1001+',
                 '11-50':'1-200',
                 '51-200':'1-200', 
                 '201-500':'201-500',
                 '5001-10000':'1001+', 
                 '1-10':'1-200', 
                 'nan':'nan'}
ds_by_applicants['no_of_employees'] = ds_by_applicants['no_of_employees'].map(employees_map)
order = ['1-200', '201-500','501-1000','1001+']
ds_by_applicants['days_from_post_date'] = np.clip(ds_by_applicants['days_from_post_date'], a_max = 58, a_min = 7)
ds_by_applicants['no_of_applicants_weekly'] = np.multiply(np.divide(ds_by_applicants['no_of_applicants'], 
                                                                    ds_by_applicants['days_from_post_date'].
                                                                    astype(int)), 7)
sns.boxplot(data = ds_by_applicants, x='no_of_applicants', 
            y='no_of_employees', order = order)
sns.stripplot(data = ds_by_applicants, x='no_of_applicants', 
            y='no_of_employees', order = order, color = 'purple')
plt.show()
```

```python
# Visualize number of applicants vs number of vacancies by industry to understand competition in each industry
my_order = ds_by_applicants.groupby('industry_new')["no_of_applicants_weekly"].median().sort_values(ascending=False).index

sns.boxplot(data = ds_by_applicants, 
            x='no_of_applicants_weekly', 
            y='industry_new',
           order = my_order)
sns.stripplot(data = ds_by_applicants, 
            x='no_of_applicants_weekly', 
            y='industry_new',
           order = my_order)
sns.stripplot(data = ds_by_industry, 
            x='count', 
            y='industry_new',
           order = my_order,
             color = 'black')
plt.show()
```

```python
# Visualize job applications by industry and size of company
sns.displot(ds_by_applicants, x='industry_new', hue = 'no_of_employees', hue_order = order, multiple = 'stack')
plt.xticks(rotation='vertical') 
plt.show()
```
